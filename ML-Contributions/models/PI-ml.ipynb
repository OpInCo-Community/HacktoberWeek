```python
from fastai.vision.all import *
```

```python
path = Path("C:\\Users\\piyus\\simple_images")
```

```python
data = ImageDataLoaders.from_folder(path,bs=8,valid_pct=0.1,item_tfms = Resize(224))
```

```python
#vision_learner
```

```python
learn = vision_learner(dls=data,arch=models.vgg11_bn,metrics=[accuracy],cbs=[ShowGraphCallback])#tranferLearning
#vision_learner?
```

```python
#epoch number of iteration 
learn.fit_one_cycle(3) #accracy 1 hai tooh overfiting
```
