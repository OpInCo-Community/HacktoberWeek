```python
from fastai.vision.all import *
```

```python
#Simple_image_download showing error so downloaded database from Kaggle(https://www.kaggle.com/datasets/gpiosenka/sports-classification?select=train)
```

```python
path=Path("/Users/jatinagrawal/Downloads/archive/train")
          
```

```python
data=ImageDataLoaders.from_folder(path,bs=16,valid_pct=0.1,item_tfms=Resize(224))
```

```python
vision_learner?
```

```python
learn = vision_learner(dls=data,arch=resnet18,metrics=[accuracy],cbs=[ShowGraphCallback()])
```

```python
learn.fit_one_cycle(3)
```

```python
from PIL import Image
```

```python
Img=Image.open("/Users/jatinagrawal/Downloads/archive/train/air hockey/1.jpg")
```

```python
import numpy as np
```

```python
learn.predict(np.array(Img))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.print_classification_report()
```

```python
interp.plot_top_losses(5)
```

```python
interp.plot_confusion_matrix(dpi=300,figsize=(14,14))
```
